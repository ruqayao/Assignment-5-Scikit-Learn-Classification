```python
#Import libraries
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

#Class to handle dataset loading
class CancerDataLoader:

    def __init__(self):
        self.data = None
        self.X = None
        self.y = None

    def load_data(self):
        dataset = load_breast_cancer()
        self.data = dataset
        self.X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
        self.y = pd.Series(dataset.target)

    def get_features(self):
        return self.X

    def get_target(self):
        return self.y


#Class to handle model training and evaluation
class ModelTrainer:

    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.X_train = None
        self.X_test = None
        self.y_train = None
        self.y_test = None
        self.scaler = StandardScaler()
    
    def split_data(self):
        split = train_test_split(self.X, self.y, test_size=0.2, random_state=42)
        self.X_train = split[0]
        self.X_test = split[1]
        self.y_train = split[2]
        self.y_test = split[3]
        #Scale the data
        self.X_train = self.scaler.fit_transform(self.X_train)
        self.X_test = self.scaler.transform(self.X_test)
    
    def train_and_evaluate(self, model, name):

        model.fit(self.X_train, self.y_train)
        predictions = model.predict(self.X_test)

        accuracy = accuracy_score(self.y_test, predictions)

        print("Model:", name)
        print("Accuracy:", accuracy)
        print("Classification Report:")
        print(classification_report(self.y_test, predictions))
        print("-----------------------------")


#Main program
def main():

    #Load dataset
    loader = CancerDataLoader()
    loader.load_data()

    X = loader.get_features()
    y = loader.get_target()

    #Prepare trainer
    trainer = ModelTrainer(X, y)
    trainer.split_data()

    #Models
    logistic = LogisticRegression(max_iter=10000)
    tree = DecisionTreeClassifier(max_depth=5)
    knn = KNeighborsClassifier(n_neighbors=7)

    #Train and evaluate
    trainer.train_and_evaluate(logistic, "Logistic Regression")
    trainer.train_and_evaluate(tree, "Decision Tree")
    trainer.train_and_evaluate(knn, "K-Nearest Neighbors")


if __name__ == "__main__":
    main()
```

```python
#Model Performance Explanation

#Three machine learning classification models were trained and evaluated: Logistic Regression, Decision Tree, and K-Nearest Neighbors (KNN). 
#The models were trained using an 80/20 train-test split and evaluated using accuracy, precision, recall, and F1-score. 
#Among the three models, Logistic Regression produced the best overall performance. 
#Logistic Regression achieved an accuracy of 0.9737, which was higher than both the Decision Tree and KNN models, which each achieved 0.9474 accuracy.

#In addition to accuracy, Logistic Regression also showed strong performance across other metrics. 
#For malignant tumors (class 0), the model achieved a precision of 0.98, recall of 0.95, and an F1-score of 0.96. 
#For benign tumors (class 1), it achieved precision of 0.97, recall of 0.99, and an F1-score of 0.98. 
#These results indicate that Logistic Regression was highly effective at correctly identifying both malignant and benign cases.

#The Decision Tree model performed reasonably well but showed slightly lower accuracy and was more prone to overfitting because it can create overly complex rules from the training data. 
#The K-Nearest Neighbors model also performed well but was slightly less consistent depending on the data distribution.
#Both models achieved 0.9474 accuracy, with precision and recall values around 0.93 for malignant tumors and 0.96 for benign tumors. 
#While these models still performed well, their slightly lower accuracy and F1-scores suggest that they were less consistent than Logistic Regression for this dataset.

#Based on the evaluation metrics, Logistic Regression provides the best balance of accuracy, precision, recall, and F1-score, making it the strongest performing model for this classification task.
```
